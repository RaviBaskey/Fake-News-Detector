```python
data_merge=pd.concat([data_fake,data_true],axis=0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data=data_merge.drop(['title','subject','date'],axis=1)
```

```python
data.isnull().sum()
```

```python
data=data.sample(frac=1)
```

```python
data.head()
```

```python
data.reset_index(inplace=True)
data.drop(['index'],axis=1,inplace=True)
```

```python
data.columns
```

```python
data.head()
```

```python

def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ", text)
    text = re.sub('https?//\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python

data['text'] = data['text'].apply(wordopt)
```

```python

x = data['text']
y = data['class']
```

```python

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25)
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python

from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train , y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python

LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test , pred_lr))
```

```python

from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train , y_train)
```

```python

pred_dt = DT.predict(xv_test)
```

```python

DT.score(xv_test, y_test)
```

```python

print(classification_report(y_test , pred_dt))
```

```python

from sklearn.ensemble import GradientBoostingClassifier

GB = GradientBoostingClassifier(random_state = 0)
GB.fit(xv_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(random_state = 0)
RF.fit(xv_train , y_train)
```

```python
pred_fr = RF.predict(xv_test)
```

```python
import pandas as pd

def output_label(n):
    if n == 0:
        return "Fake news"
    elif n == -1:
        return "Not A fake News"
    else:
        return "Unknown prediction"

def manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)  # Ensure wordopt is defined
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)  # Ensure vectorization is defined

    pred_LR = LR.predict(new_xv_test)  # Ensure LR is defined
    pred_DT = DT.predict(new_xv_test)  # Ensure DT is defined
    pred_GB = GB.predict(new_xv_test)  # Ensure GB is defined
    pred_RF = RF.predict(new_xv_test)  # Ensure RF is defined

    predictions = (
        output_label(pred_LR[0]),
        output_label(pred_DT[0]),
        output_label(pred_GB[0]),
        output_label(pred_RF[0])
    )

    return "\n\nLR Prediction: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(*predictions)
```
